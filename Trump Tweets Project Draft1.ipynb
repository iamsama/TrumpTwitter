```python
tweets = pd.read_csv("tweets.csv")
tweets.to_csv("tweets_copy.csv", index=False)
```

```python
tweets = tweets.set_index('source')
```

```python
tweets = tweets.drop('Twitter Ads', axis=0)
```

```python
tweets.text = tweets.text.str.replace('&amp;', '&')
```

```python
# Most retweetted tweets:
tweets = tweets.sort_values(['retweet_count'], ascending=False)
tweets.head(20)
```

```python
# Trump tweets' retweet total:
tweets.retweet_count.sum()
```

```python
# Trump tweets' retweet average:
tweets.retweet_count.mean()
```

```python
# Most favourite tweets:
tweets = tweets.sort_values(['favorite_count'], ascending=False)
tweets.head(20)
```

```python
# Trump tweets' favourite total:
tweets.favorite_count.sum()
```

```python
# Trump tweets' favourite average:
tweets.favorite_count.mean()
```

```python
# How many of Trump tweets are actually retweets?
tweets.is_retweet.value_counts()
```

```python
# Trump's favourite tweet time during the day (week, month, year, all-time):
# tweets.created_at.value_counts().idxmax()
tweet_hours = pd.to_datetime(tweets.created_at)
tweet_hours = tweet_hours.dt.strftime('%H:%M')
tweet_hours.value_counts().idxmax()
```
